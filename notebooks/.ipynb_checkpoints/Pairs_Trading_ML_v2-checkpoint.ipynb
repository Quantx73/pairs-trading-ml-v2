```python
# Baseline Strategy: Z-Score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Load data
data_path = Path('../data/etf_pairs.csv')
prices = pd.read_csv(data_path, index_col=0, parse_dates=True)

# Parameters
window = 60
entry_threshold = 2.0
exit_threshold = 0.5
transaction_cost = 0.001

# Calculate ratio and z-score
prices['ratio'] = prices['EWA'] / prices['EWC']
prices['ratio_ma'] = prices['ratio'].rolling(window).mean()
prices['ratio_std'] = prices['ratio'].rolling(window).std()
prices['zscore'] = (prices['ratio'] - prices['ratio_ma']) / prices['ratio_std']

# Generate signals
prices['position'] = 0
prices['signal'] = 0

prices.loc[prices['zscore'] < -entry_threshold, 'signal'] = 1   # Long spread
prices.loc[prices['zscore'] > entry_threshold, 'signal'] = -1   # Short spread

# Simulate positions
position = 0
for i in range(len(prices)):
    signal = prices.iloc[i]['signal']
    
    if position == 0:
        if signal == 1:
            position = 1
        elif signal == -1:
            position = -1
    elif position == 1:
        if abs(prices.iloc[i]['zscore']) < exit_threshold or signal == -1:
            position = 0
    elif position == -1:
        if abs(prices.iloc[i]['zscore']) < exit_threshold or signal == 1:
            position = 0
    
    prices.iloc[i, prices.columns.get_loc('position')] = position

# Calculate returns
prices['return_A'] = prices['EWA'].pct_change()
prices['return_B'] = prices['EWC'].pct_change()
prices['strategy_return'] = prices['position'] * (prices['return_A'] - prices['return_B'])
prices['position_change'] = prices['position'].diff().abs() / 2
prices['strategy_return_net'] = prices['strategy_return'] - (prices['position_change'] * transaction_cost)
prices['cumulative_return'] = (1 + prices['strategy_return_net']).cumprod()

# Results
print("=" * 50)
print("BASELINE STRATEGY RESULTS")
print("=" * 50)
print(f"Window: {window} days")
print(f"Total Return (net): {(prices['cumulative_return'].iloc[-1] - 1):.2%}")
print(f"Number of trades: {(prices['position'].diff() != 0).sum() / 2:.0f}")
print(f"Win Rate: {(prices['strategy_return'] > 0).mean() * 100:.1f}%")
```

```python
# Plot results
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Z-score with signals
axes[0].plot(prices.index, prices['zscore'], color='blue', linewidth=1)
axes[0].axhline(y=entry_threshold, color='red', linestyle='--', label='Entry +2')
axes[0].axhline(y=-entry_threshold, color='green', linestyle='--', label='Entry -2')
axes[0].axhline(y=exit_threshold, color='orange', linestyle=':', label='Exit ±0.5')
axes[0].axhline(y=-exit_threshold, color='orange', linestyle=':')
axes[0].set_title('Z-Score and Trading Signals')
axes[0].set_ylabel('Z-Score')
axes[0].legend(loc='upper right')
axes[0].grid(True, alpha=0.3)

# Cumulative returns
axes[1].plot(prices.index, prices['cumulative_return'], color='green', linewidth=2)
axes[1].axhline(y=1, color='gray', linestyle='--', alpha=0.5)
axes[1].set_title('Cumulative Returns (Net of Transaction Costs)')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Cumulative Return')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Feature Engineering for ML
import pandas as pd
import numpy as np
from pathlib import Path

# Load data
data_path = Path('../data/etf_pairs.csv')
prices = pd.read_csv(data_path, index_col=0, parse_dates=True)

# Parameters
window = 60
lags = [1, 2, 3, 5]

# Calculate ratio and z-score
prices['ratio'] = prices['EWA'] / prices['EWC']
prices['ratio_ma'] = prices['ratio'].rolling(window).mean()
prices['ratio_std'] = prices['ratio'].rolling(window).std()
prices['zscore'] = (prices['ratio'] - prices['ratio_ma']) / prices['ratio_std']

# Returns
prices['return_A'] = prices['EWA'].pct_change()
prices['return_B'] = prices['EWC'].pct_change()
prices['return_spread'] = prices['return_A'] - prices['return_B']

# Volatility (annualized)
prices['vol_A'] = prices['return_A'].rolling(window).std() * np.sqrt(252)
prices['vol_B'] = prices['return_B'].rolling(window).std() * np.sqrt(252)
prices['vol_ratio'] = prices['vol_A'] / prices['vol_B']

# Rolling correlation
prices['corr'] = prices['return_A'].rolling(window).corr(prices['return_B'])

# Create lagged features
for lag in lags:
    prices[f'zscore_lag_{lag}'] = prices['zscore'].shift(lag)
    prices[f'return_A_lag_{lag}'] = prices['return_A'].shift(lag)
    prices[f'return_B_lag_{lag}'] = prices['return_B'].shift(lag)

# Create target variable (mean reversion within 5 days)
horizon = 5
prices['future_zscore'] = prices['zscore'].shift(-horizon)
prices['target'] = ((abs(prices['future_zscore']) < abs(prices['zscore'])) & 
                    (abs(prices['zscore']) > 1.5)).astype(int)

# Drop NaN values
features = prices.dropna()

print("=" * 50)
print("FEATURE ENGINEERING")
print("=" * 50)
print(f"Total samples: {len(features)}")
print(f"Target distribution:")
print(f"  No trade (0): {(features['target'] == 0).sum()} ({(features['target'] == 0).mean()*100:.1f}%)")
print(f"  Trade (1): {(features['target'] == 1).sum()} ({(features['target'] == 1).mean()*100:.1f}%)")S
```
